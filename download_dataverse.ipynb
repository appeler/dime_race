```python
import re
import urllib.request

import requests
from dataverse import Connection
from tqdm import tqdm
import pandas as pd
```

```python
class DownloadProgressBar(tqdm):
    def update_to(self, b=1, bsize=1, tsize=None):
        if tsize is not None:
            self.total = tsize
        self.update(b * bsize - self.n)

def download_url(url, output_path):
    with DownloadProgressBar(unit='B', unit_scale=True,
                             miniters=1, desc=url.split('/')[-1]) as t:
        urllib.request.urlretrieve(url, filename=output_path, reporthook=t.update_to)
```

```python
r = requests.get('https://dataverse.harvard.edu/api/datasets/:persistentId/?persistentId=doi:10.7910/DVN/O5PX0B')
j = r.json()
j
```

```python
ts = 0
for f in j['data']['latestVersion']['files']:
    fn = f['dataFile']['filename']
    m = re.match('contribDB_(\d+).*', fn)
    if m:
        fs = f['dataFile']['filesize']
        pid = f['dataFile']['persistentId']
        print(fn, fs, pid)
        ts += fs
        if fn.startswith('contribDB_19') or fn == 'contribDB_2000.csv.gz':
            url = 'http://dataverse.harvard.edu/api/access/datafile/:persistentId/?persistentId=%s' % pid
            ofn = '/opt/data/dime-race/%s' % fn
            download_url(url, ofn)
print(ts/(1024*1024))
```
